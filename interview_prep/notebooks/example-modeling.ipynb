```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

HOME_DIR = os.path.expanduser("~")
DATA_DIR = os.path.join(HOME_DIR, 'git', 'hello-world', 'interview_prep', 'data')
```

```python
def check_missing_values(data):
    """Checks Missing Values By Column in Data"""
    df_count_missing = pd.DataFrame({
        'number_total_rows': data.shape[0],
        'number_missing_rows': data.isnull().sum(),
        'percentage': data.isnull().mean()})
    return df_count_missing
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, 'cfa_exercise_data.csv'))
```
